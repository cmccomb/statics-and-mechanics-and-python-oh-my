```
# Import the scipy library.
# We really just use this library for one thing - an accurate decimal value for g.
import scipy
```

```
# Import the numpy library.
# This library is useful for things that have to do with linear algebra.
import numpy
```

```
# Define some constants
mass_of_flowerpot = 10
g = 10
```

```
g
```

```
# Define the locations of each attachment point
A = numpy.array([1.5, 2.0, 0.0])
B = numpy.array([1.5, 3.0, 0.0])
C = numpy.array([2.5, 2.0, 0.0])
D = numpy.array([0.0, 0.0, 1.5])
```

```
# Next, define a lambda function to help us calculate unit vectors
unit_vector = lambda x, y: (x-y)/numpy.linalg.norm(x-y);
```

```
# Calculate those unit vectors
uAB = unit_vector(B, A)
uAC = unit_vector(C, A)
uAD = unit_vector(D, A)
```

```
uAD
```

```
# Define the linear system of equations, start with X
X = numpy.transpose(numpy.stack([uAB, uAC, uAD]))
X
```

```
# Y is the external force vector
Y = numpy.array([0, 0, mass_of_flowerpot*g])
Y
```

```
# Solve for the forces
numpy.linalg.solve(X, Y)
```
